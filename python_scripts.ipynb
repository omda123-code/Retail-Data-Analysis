```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
payments=pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/Brazilian E-Commerce Public Dataset by Olist/data_export/top_payment_type.csv')
payments
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=payments,x='payment_type',y='totalPayment')
plt.title('Top Payment Method')
plt.xlabel('Payment Type')
plt.ylabel('Total Revenue')
plt.savefig('top_payment_type.png',bbox_inches='tight')
plt.show()
```

```python
cities=pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/Brazilian E-Commerce Public Dataset by Olist/data_export/orders_per_city.csv')
```

```python
top_cities=cities.sort_values('totalOrders',ascending=False).head(10)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=top_cities,x='customer_city',y='totalOrders')
plt.title('Top 10 Cities by Number of Orders')
plt.xticks(rotation=45)
plt.xlabel('City')
plt.ylabel('Total Orders')
plt.savefig('orders_per_city.png',bbox_inches='tight')
plt.show()
```

```python
Late=pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/Brazilian E-Commerce Public Dataset by Olist/data_export/Late_delivery_percentage.csv')
```

```python
late_summary = Late.groupby('order_status')['late_delivery_percentage'].sum().reset_index()

plt.figure(figsize=(6,6))
plt.pie(
    late_summary['late_delivery_percentage'],
    labels=late_summary['order_status'],
    autopct='%1.1f%%'
)
plt.title('Late vs On-time Orders')
plt.savefig('Late_delivery_percentage.png', bbox_inches='tight')
plt.show()
```

```python
status=pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/Brazilian E-Commerce Public Dataset by Olist/data_export/orders_status_percentage.csv')
```

```python
status_summary=status.groupby('order_status')['count_orders'].sum().reset_index()
plt.figure(figsize=(6,6))
plt.pie(
   status_summary['count_orders'],labels=status_summary['order_status'],autopct='1%.1f%%')
plt.title('Order Status Distribution')
plt.savefig('orders_status_percentage.png',bbox_inches='tight')
plt.show()
```
